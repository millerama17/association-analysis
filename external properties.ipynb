```python
import pandas as pd
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
query_ep = """
SELECT ?property ?propertyLabel WHERE {
  ?property wdt:P31 wd:Q18614948;  # Instance of "external identifier"
          rdfs:label ?propertyLabel .
  FILTER(LANG(?propertyLabel) = "en")
}
ORDER BY ?property
"""
```

```python
sparql_endpoint = "https://query.wikidata.org/sparql"
sparql = SPARQLWrapper(sparql_endpoint)
```

```python
sparql.setQuery(query_ep)
sparql.setReturnFormat(JSON)
results_ep = sparql.query().convert()
```

```python
codes = [item['property']['value'].split('/')[-1] for item in results_ep['results']['bindings']]
property_labels = [item['propertyLabel']['value'] for item in results_ep['results']['bindings']]
```

```python
df = pd.DataFrame({
    'Code': codes,
    'Property Label': property_labels
})
```
